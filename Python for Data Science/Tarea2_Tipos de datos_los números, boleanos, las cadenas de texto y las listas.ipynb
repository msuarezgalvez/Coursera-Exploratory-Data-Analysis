### Los Números

Dentro de los numeros nos encontramos los enteros, los decimales y los números 
complejos. Estos se definen como int, float y complex, respectivamente.

Podemos utilizar la función type para conocer el tipo de dato al que pertenece una variable. Y también podemos convertir un tipo a otro con los métodos:
- int( )
- float( )
- complex( )


```python
x = 4    # Entero
y = 6.6  # Decimal
z = 7j   # Complejo

a = 5
print(a, "es del tipo", type(a))

a = 2.0
print(a, "es del tipo", type(a))

#convertir un entero a un decimal
a = float(x)
print(type(a))

#convertir un decimal a un entero
b = int(y)
print(type(b))
```

Se pueden realizar múltiples operaciones sobre los números como: 
- Sumar + 
- Restar - 
- Multiplicar * 
- Dividir /
- Elevar al exponente **

También hay funciones como random ( ) para generar números aleatorios, range ( ) para generar números en un rango, etc.

```python
# establecemos las variables x e y
x = 4    
y = 6.6 

# suma de x e y
z = x + y
print('la suma de 4 y 6.6 es:',z)

import random
print('un número aleatorio entre 1 y 50 es:',random.randrange(1, 50))
```

### Booleanos


Los booleanos son un tipo de datos del tipo True o False. Se pueden utilizar para generar condicionales. Los condicionales se ejecutan cuando es igual a True. Pueden combinarse diferentes condicionales con: and, or, not o ‘==’, ‘!=’, ‘>’, ‘<=’.

```python
x = 4
y = 1

if x > y:
    print('x mayor que y')# se ejecuta la condicion que es verdad (True)
else:
    print('menos que')

## Como 5 es diferente de 4 devuelve False    
print(5==4)
print(5 is 4)

# Como ambos son el decimal de 5 devuelve True
print(float(5.0) == float(5))
```

### Cadenas de texto

Son un conjunto de arrays de carácteres inmutables. Estos están delimitados por comillas ' ' o por comillas dobles " ". Los carácteres del tipo ‘\n’ indican un salto de linea o ‘\t’ tabulación.


Al igual que en las listas y tuplas, se puede seleccionar un subconjunto de carácteres mediante [ ]. También podemos aplicarles diferentes métodos del tipo len(String), upper( ) y lower( ).Con format( ) podremos añadir un valor de manera dinámica.

```python
## generamos el string Hola Mundo
palabra = 'Hola Mundo'
print(palabra)

#Seleccionamos un subvconjunto de carácteres
print(palabra[3])
print(palabra [0:2])

# Aplicamos los métodos de len() y  de upper()
print('lalongitud es',len(palabra))
palabra.upper()

# Aplicamos el método format
print('la palabra seleccionada es {}'.format(palabra))
palabra = 'Adios'
print('la palabra seleccionada es {}'.format(palabra))
```

### Listas

Las listas son un conjunto de elementos ordenados que pueden ser de igual o diferente tipo. Se generan con el método list( ). Al igual que con las cadenas de texto se puede utilizar [ ] para acceder a un subconjunto de elementos. Tienen diferentes métodos como .append( ), .pop( ), .index( ), etc.

```python
# se generan las listas numeros y elementos
numeros = [2, 3, 5, 7]
print(numeros)
elementos = ['coche', 2, 3,'casa', False]
print(elementos)

# selección de un subconjunto de elementos
print(elementos[0])
print(elementos[1:3])

sorted(numeros)
sum(numeros)

numerosAltos = [ ]
for i in numeros:
    if i > 3:
        numerosAltos.append(i)
    else:
        pass

print(numerosAltos)
```
